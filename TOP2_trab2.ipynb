# IMPORTS

```python
import cv2               as cv
import matplotlib.pyplot as plt
import numpy             as np
```

# TRATAMENTO

```python
img = cv.imread( '/content/lena512.bmp', 0 )
edges = cv.Canny( img, 120, 240 )

kernel = np.ones( ( 5, 5 ), np.uint8 )
dilation = cv.dilate(edges, kernel, iterations = 1)  #AQUI SE ESCOLHE O TAMANHO DA BORDA
```

# PLOT IMAGEM

## Delimitação de borda da imagem

```python
fig, ax = plt.subplots( ncols = 3, figsize=(15,5) )
ax[0].imshow( img, cmap = 'gray' )                     # AQUI SE ESCOLHE A COR
ax[0].set_title( 'Imagem original' ) 
ax[0].axis( 'off' )
ax[1].imshow( edges, cmap = 'magma' )
ax[1].set_title( 'Imagem com borda' )
ax[1].axis( 'off' )
ax[2].imshow( dilation, cmap = 'gray' )                # AQUI SE ESCOLHE A COR
ax[2].set_title( 'Imagem com borda reforçada' )
plt.axis( 'off' )
plt.show()
```

## Adicionando moldura a imagem

```python
from google.colab.patches import cv2_imshow
import cv2 
   
# caminho 
path = '/content/lena512.bmp'
image = cv2.imread(path)
    
start_point = (0, 0)
end_point = (510, 510)
  
color = (0, 0, 0)            #AQUI SE ESCOLHE A COR

thickness = 50               #AQUI SE ESCOLHE A GROSSURA DA MOLDURA
 
image = cv2.rectangle(image, start_point, end_point, color, thickness)
cv2_imshow(image) 
```
